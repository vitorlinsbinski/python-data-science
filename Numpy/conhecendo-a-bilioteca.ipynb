# Transformar lista em array

### Benefícios de Utilizar Arrays ao Invés de Listas no Python

1. **Eficiência de Processamento**  
   - Arrays são otimizados para trabalhar com grandes conjuntos de dados homogêneos.  
   - Liberação de memória ocorre de forma mais eficiente, reduzindo o consumo de recursos.

2. **Facilidade de Uso**  
   - Permitem operações matemáticas mais claras e concisas.  
   - Exemplo: somar dois arrays diretamente, sem necessidade de loops.

3. **Integração com Outras Bibliotecas**  
   - Arrays são amplamente utilizados em bibliotecas populares como *pandas* e *matplotlib*.  
   - Facilitam análises de dados e visualizações devido à compatibilidade nativa.


```python
import numpy as np

lista = [1,2,3,4,5]

array = np.array(lista)

print("Lista: ", lista)
print("Array: ", array)
```

### Teste de eficiência

```python
import numpy as np
import time

# cria uma lista com 1000000 elementos
lista = list(range(1000000))

# transforma a lista em um array Numpy
array = np.array(lista)

# começa a cronometrar o tempo para a operação com a lista
start_time = time.time()

# realiza a operação de elevar ao quadrado cada elemento da lista
lista_quadrado = [i**2 for i in lista]

# para o cronômetro
list_time = time.time() - start_time

# começa a cronometrar o tempo para a operação com o array
start_time = time.time()

# realiza a operação de elevar ao quadrado cada elemento do array
array_quadrado = array**2

# para o cronômetro
array_time = time.time() - start_time

print("Tempo da operação com a lista: ", list_time)
print("Tempo da operação com o array: ", array_time)
```

# Multiplicação

Fórmula utilizada para calcular o coeficiente angular (â) de uma reta de regressão simples, onde o modelo é da forma: 

Y = aX + b

Assim, ela calcula a inclinação da reta no método dos mínimos quadrados, que minimiza a soma dos quadrados dos erros (distâncias verticais entre os pontos reais e os estimados pela reta).

Ela aparece com frequência em análises estatísticas e machine learning para modelar relações lineares entre variáveis.

O objetivo é encontrar a melhor reta que aproxima todos os pontos da curva, minimizando a soma dos quadrados das distâncias

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
X**2
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y)) / (n*np.sum(X**2) - (np.sum(X))**2)

a
```

Coeficiente linear(b)



```python
b = np.mean(Y) - a * np.mean(X)

b
```

```python
y = a*X + b
```

```python
np.linalg.norm(Moscow - y)
```

```python
plt.plot(datas, y)
plt.plot(datas, Moscow)
```
