```python
dados_apartamento = pd.read_json('dados_locacao.json')

dados_apartamento.head()
```

```python
dados_nome_apartamento = []
dados_datas_combinadas_pagamento = []
dados_data_de_pagamento = []
dados_valor_aluguel = []

for locacao in dados_apartamento['dados_locacao']:
    for data_combinada_pagamento, data_de_pagamento, valor_aluguel in zip(locacao['datas_combinadas_pagamento'], locacao['datas_de_pagamento'], locacao['valor_aluguel']):
        dados_nome_apartamento.append(locacao['apartamento'])
        dados_datas_combinadas_pagamento.append(data_combinada_pagamento)
        dados_data_de_pagamento.append(data_de_pagamento)
        dados_valor_aluguel.append(valor_aluguel)

dados_apartamento = pd.DataFrame({
    'apartamento': dados_nome_apartamento,
    'data_combinada_pagamento': dados_datas_combinadas_pagamento,
    'data_pagamento': dados_data_de_pagamento,
    'valor_aluguel': dados_valor_aluguel
})

dados_apartamento.head()
```

```python
dados_apartamento['apartamento'] = dados_apartamento['apartamento'].apply(lambda x: x.replace('(blocoAP)', '').strip())
dados_apartamento['data_combinada_pagamento'] = pd.to_datetime(dados_apartamento['data_combinada_pagamento'], format='%d/%m/%Y')
dados_apartamento['data_pagamento'] = pd.to_datetime(dados_apartamento['data_pagamento'], format='%d/%m/%Y')
dados_apartamento['valor_aluguel'] = dados_apartamento['valor_aluguel'].apply(lambda x: x.replace('$','').replace('reais', '').replace(',', '').strip())
dados_apartamento['valor_aluguel'] = dados_apartamento['valor_aluguel'].astype(np.float64)

print(dados_apartamento.info())

dados_apartamento.head()
```

```python
dados_apartamento['pagamento_atrasado'] = (dados_apartamento['data_pagamento'] - dados_apartamento['data_combinada_pagamento']).dt.days > 0

dados_apartamento.head()
```

```python
resultado = dados_apartamento.groupby('apartamento')['pagamento_atrasado'].apply(lambda x: ((x == False).sum() / len(x)) * 100).sort_values(ascending=False)
resultado
```

```python
dados_vendas = pd.read_json('dados_vendas.json')

dados_data_de_venda = []
dados_cliente = []
dados_valor_da_compra = []

for venda in dados_vendas['dados_vendas']:
    for cliente, valor_da_compra in zip(venda['Cliente'], venda['Valor da compra']):
        dados_data_de_venda.append(venda['Data de venda'])
        dados_cliente.append(cliente)
        dados_valor_da_compra.append(valor_da_compra)

dados_vendas = pd.DataFrame({
    'data_de_venda': dados_data_de_venda,
    'cliente': dados_cliente,
    'valor_da_compra': dados_valor_da_compra
})

dados_vendas.head()
```

```python
dados_vendas['data_de_venda'] = pd.to_datetime(dados_vendas['data_de_venda'], format='%d/%m/%Y')
dados_vendas['cliente'] = dados_vendas['cliente'].str.replace(r'[^a-zA-Z-À-ÿ]', ' ', regex=True).str.replace('\s+', ' ', regex=True).str.strip().str.lower().str.title()
dados_vendas['valor_da_compra'] = dados_vendas['valor_da_compra'].str.replace('R$', '').str.replace('.', '').str.replace(',', '.').str.strip().astype(np.float64)

dados_vendas.head()
```

```python
total_compras_por_cliente = dados_vendas.groupby('cliente')['valor_da_compra'].sum()
cliente_com_maior_gasto = total_compras_por_cliente.idxmax()
valor_total_maior_gasto = total_compras_por_cliente.max()

print(f'O(a) cliente {cliente_com_maior_gasto} foi o que mais gastou, totalizando um valor de R${valor_total_maior_gasto}')
```
