```python
import pandas as pd
```

```python
df = pd.read_csv('banco_clientes.csv')

df.head()
```

```python
df['nome'] = df['nome'].str.upper()

df.head()
```

```python
df['telefone'] = df['telefone'].str.replace(r'[^0-9\-]', ' ', regex=True)

df.head()
```

```python
df['cep'] = df['endereco'].str.extract(r'(\d{5}-\d{3})')
df['endereco'] = df['endereco'].str.replace(r'(\d{5}-\d{3})', '', regex=True).str.strip()

df.head()
```

```python
df[['rua', 'numero']] = df['endereco'].str.replace(',', '').str.split(r'\s(?=\d+$)', expand=True)
df.drop(columns='endereco', inplace=True)

df.head()
```
